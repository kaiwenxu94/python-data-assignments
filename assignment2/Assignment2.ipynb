```python
df = pd.read_csv('journalistskilled.csv')
```

```python
year=[]
for i in range(0,1324):
    individualyear=df['date'][i][-4:]
    year.append(individualyear)
df['years']=year
df
```

```python
country=[]
for i in range(0,1324):
    individualcountry=df['location'][i].split()[-1]
    country.append(individualcountry)
country
df['country']=country
df
```

```python
dictallcountries=df['country'].value_counts().to_dict()
countryname=list(dictallcountries)
ctcode = pd.read_csv('country code.csv')
for word in countryname:
    for k in range(len(ctcode)):
        if word==ctcode['Country'][k]:
            n=countryname.index(word)
            countryname.remove(word)
            countryname.insert(n,ctcode['code'][k])
countryname
```

```python
dforganization=df['organization'].replace({'freelance':'Freelance','Freelancer':'Freelance'}).value_counts()
dforganization
```

```python
len(dforganization)
```

```python
number_of_formal_staff=0
for i in range(1,len(dforganization)):
    number_of_formal_staff=number_of_formal_staff+dforganization[i]
print(number_of_formal_staff)
```

```python
freelancer_vs_staff = pd.Series([dforganization['Freelance'],number_of_formal_staff], index=['Number of Freelancers', 'Number of Formal Staff'])
freelancer_vs_staff
```

```python
ax3=freelancer_vs_staff.plot.pie(figsize=(6, 6),autopct='%1.1f%%',shadow=True, startangle=90)
ax3.set_title('Number of Killed Freelancers vs Formal Staff')
```
