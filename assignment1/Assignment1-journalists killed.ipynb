```python
from selenium import webdriver
import time

```

```python
def get_journalists_from_browser(browser):
    journalists = []
    for session in browser.find_elements_by_css_selector("tbody tr"): 
        journalist = {}
        h = session.find_element_by_css_selector("td a")
        journalist['name'] = h.text 
        journalist['url'] = h.get_attribute('href')
        journalist['organization']=session.find_element_by_css_selector("td:nth-of-type(2)").text
        journalist['date']=session.find_element_by_css_selector("td:nth-of-type(3)").text
        journalist['location']=session.find_element_by_css_selector("td:nth-of-type(4)").get_attribute('title')
        journalist['type of death']=session.find_element_by_css_selector("td[class='js-column-type']").text
        journalists.append(journalist)
    return journalists
```

```python
url="https://cpj.org/data/killed/?status=Killed&motiveConfirmed%5B%5D=Confirmed&type%5B%5D=Journalist&start_year=1992&end_year=2018&group_by=year"
browser = webdriver.Chrome()
browser.get(url)
time.sleep(2)
```

```python
all_journalists_killed = []
for i in range(67):
    time.sleep(7)
    try:
        journalists_each_page = get_journalists_from_browser(browser)
        all_journalists_killed.extend(journalists_each_page)
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight/1.5);')
        next_page = browser.find_element_by_link_text('Next')
        next_page.click()
    except Exception as e:
        print(e)
        print('Error on page %s' % i)
```

```python
import pandas as pd 
df = pd.DataFrame(all_journalists_killed) 
df
```
